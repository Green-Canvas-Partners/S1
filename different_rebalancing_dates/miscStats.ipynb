```python
import numpy as np
from collections import defaultdict

def exponential_weights(*, length, alpha=0.7):
    """
    Generate exponential weights for a given length.

    Args:
        length (int): Number of weights to generate.
        alpha (float): Decay factor for the weights, typically between 0 and 1.

    Returns:
        np.array: Array of weights, where weights exponentially decrease.
    """
    indices = np.arange(length)
    weights = alpha ** indices
    weights /= weights.sum()  # Normalize weights to sum to 1
    return weights  # Return weights in reverse order

# def calculate_portfolio_weights_by_date(strategies):
#     """
#     Calculate the weight of each stock in the portfolio at each date across multiple strategies.

#     Args:
#         strategies (list of dict): List of strategies, where each strategy is a dictionary
#                                    with dates as keys and arrays of stocks as values.

#     Returns:
#         dict: A dictionary with dates as keys and dictionaries of stock weights as values.
#     """
#     portfolio_weights_by_date = defaultdict(dict)
#     n = len(strategies)

#     # Get all unique dates across all strategies
#     all_dates = set()
#     for strategy in strategies:
#         all_dates.update(strategy.keys())
#     all_dates = sorted(all_dates)  # Sort dates chronologically

#     for date in all_dates:
#         date_weights = defaultdict(float)

#         # Total stock weight to distribute
#         total_stock_weight = 0.0

#         for strategy in strategies:
#             if date in strategy:
#                 stocks = strategy[date]
#                 if len(stocks) > 0:
#                     weights = exponential_weights(length=len(stocks))
#                     weights /= n  # Normalize by the number of strategies
#                     for stock, weight in zip(stocks, weights):
#                         date_weights[stock] += weight
#                         total_stock_weight += weight

#         # Calculate cash weight to ensure total weights sum to 1
#         cash_weight = 1.0 - total_stock_weight

#         # Add cash weight to the portfolio for this date
#         date_weights['CASH'] = cash_weight

#         # Store the weights for this date
#         portfolio_weights_by_date[date] = dict(date_weights)

#     return dict(portfolio_weights_by_date)

def calculate_portfolio_weights_by_date(strategies, strategy_weights):
    """
    Calculate the weight of each stock in the portfolio at each date across multiple strategies.

    Args:
        strategies (list of dict): List of strategies, where each strategy is a dictionary
                                   with dates as keys and arrays of stocks as values.
        strategy_weights (list of float): List of weights for each strategy, where the sum
                                          of these weights should be equal to 1.

    Returns:
        dict: A dictionary with dates as keys and dictionaries of stock weights as values.
    """
    # Ensure that strategy_weights sum to 1.0 (if they don't, normalize them)
    total_weight = sum(strategy_weights)
    if total_weight != 1.0:
        strategy_weights = [weight / total_weight for weight in strategy_weights]

    portfolio_weights_by_date = defaultdict(dict)
    n = len(strategies)

    # Get all unique dates across all strategies
    all_dates = set()
    for strategy in strategies:
        all_dates.update(strategy.keys())
    all_dates = sorted(all_dates)  # Sort dates chronologically

    for date in all_dates:
        date_weights = defaultdict(float)

        # Total stock weight to distribute
        total_stock_weight = 0.0

        for strategy_idx, strategy in enumerate(strategies):
            if date in strategy:
                stocks = strategy[date]
                if len(stocks) > 0:
                    weights = exponential_weights(length=len(stocks))
                    weights *= strategy_weights[strategy_idx]  # Scale by strategy weight
                    for stock, weight in zip(stocks, weights):
                        date_weights[stock] += weight
                        total_stock_weight += weight

        # Calculate cash weight to ensure total weights sum to 1
        cash_weight = 1.0 - total_stock_weight

        # Add cash weight to the portfolio for this date
        date_weights['CASH'] = cash_weight

        # Store the weights for this date
        portfolio_weights_by_date[date] = dict(date_weights)

    return dict(portfolio_weights_by_date)



all=[]
for i in range(18):
    strategies = [ss['1'][i], ss['2'][i], ss['3'][i], ss['4'][i], ss['5'][i], ss['6'][i], ss['7'][i]]#ss['1'][i], ss['2'][i], ss['3'][i], ss['4'][i], ss['5'][i], ss['6'][i], ss['7'][i]

    portfolio_weights_by_date = calculate_portfolio_weights_by_date(strategies, [0.225, 0.05, 0.15, 0.1, 0.2, 0.225, 0.05])#[0.225, 0.05, 0.15, 0.1, 0.2, 0.225, 0.05]

    # Output portfolio weights by date
    all.append(portfolio_weights_by_date)


from collections import defaultdict

# Function to convert date keys to year-month format
def convert_to_year_month(date_dict):
    new_dict = defaultdict(dict)
    for date, weights in date_dict.items():
        year_month = date[:7]
        new_dict[year_month].update(weights)
    return dict(new_dict)

# Apply the conversion to each dictionary in the 'all' list
all_year_month = [convert_to_year_month(strategy) for strategy in all]

# Step 1: Equally divide the weights for each strategy
strategies = [all_year_month[i] for i in range(len(all_year_month))]

# Divide weights by len(all_year_month))] for each strategy
for strategy in strategies:
    for date in strategy:
        for ticker in strategy[date]:
            strategy[date][ticker] /= len(all_year_month)

# Step 2: Aggregate the weights for each mutual month
aggregated_weights = {}

# Find common months between all strategies
common_dates = set(strategies[0].keys())
for strategy in strategies[1:]:
    common_dates = common_dates.intersection(set(strategy.keys()))

for date in common_dates:
    aggregated_weights[date] = {}
    # Get all unique tickers for the current date
    all_tickers = set()
    for strategy in strategies:
        all_tickers.update(strategy[date].keys())
    
    for ticker in all_tickers:
        # Sum the weights from all strategies, defaulting to 0 if the ticker is not present
        total_weight = sum(strategy[date].get(ticker, 0) for strategy in strategies)
        aggregated_weights[date][ticker] = total_weight

# Sort the aggregated weights by date
aggregated_weights = dict(sorted(aggregated_weights.items()))

# Sort the values in the dict by weights
for date in aggregated_weights:
    aggregated_weights[date] = dict(sorted(aggregated_weights[date].items(), key=lambda item: item[1], reverse=True))

aggregated_weights



```

```python
# Threshold for filtering

import copy

data=copy.deepcopy(aggregated_weights)
threshold = 0.025

# Process the data
for month, weights in data.items():
    others_weight = 0.0
    filtered_weights = {}
    
    for ticker, weight in weights.items():
        if weight > threshold:
            filtered_weights[ticker] = weight
        else:
            others_weight += weight
    
    # Add the OTHERS ticker if there are any weights below the threshold
    if others_weight > 0:
        filtered_weights['OTHERS'] = others_weight
    
    # Update the original dictionary with the filtered weights
    data[month] = filtered_weights

data
```
